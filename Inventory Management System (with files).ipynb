# Product Details
1) Product id
2) Product name
3) Product price
4) Quantity

```python
import time

#Opening and reading the inventory
fd=open("Inventory.txt","r")
products=fd.read().split("\n")
fd.close()

#Getting the input from the user
user_username=input("Enter username: ")
user_email=input("Enter email: ")
user_phoneno=input("Enter mobile number: ")

user_prodId=input("Enter product Id: ")
user_prodqty=input("Enter quantity: ")

updated_productList=[]


for product in products:
    product_details=product.split(",")
    
    #checking if the product exists in inventory
    if(user_prodId==product_details[0]):
        #checking if the entered quantity exists
        if(int(user_prodqty) <= int(product_details[3])):
            print("**************************************************")
            print("Product ID:               ",product_details[0])
            print("Product name:             ",product_details[1])
            print("Price:                    ",product_details[2])
            print("Quantity:                 ",product_details[3])
            print("**************************************************")
            print("Total amount:             ",int(user_prodqty)*int(product_details[2]))

            product_details[3]=str(int(product_details[3])-int(user_prodqty))
            
            #Generating sales
            fd=open("Sales.txt","a")
            sales_details = user_username+","+user_email+","+user_phoneno+","+user_prodId+","+product_details[1]+","+str(int(user_prodqty)*int(product_details[2]))+","+time.ctime()+"\n"
            
            
            fd.write(sales_details)
            fd.close()
            
        #if not having the required quantity
        else:
            print("Sorry, the avaiable quantity is: ",product_details[3])
            print("Do you wish to proceed with the whole available quantity?")
            
            ch=input("Enter y/n: ")
            if(ch=="Y" or ch=="y"):
                print("**************************************************")
                print("Product ID:               ",product_details[0])
                print("Product name:             ",product_details[1])
                print("Price:                    ",product_details[2])
                print("Quantity:                 ",product_details[3])
                print("**************************************************")
                print("Total amount:             ",int(product_details[3])*int(product_details[2]))
                
                 #Generating sales
                fd=open("Sales.txt","a")
                sales_details = user_username+","+user_email+","+user_phoneno+","+user_prodId+","+product_details[1]+","+str(int(product_details[3])*int(product_details[2]))+","+time.ctime()+"\n"

                fd.write(sales_details)
                fd.close()
                
                product_details[3]="0"
            else:
                print("Thankyou!")

    #updating the inventory string
    updated_productList.append(product_details)
    
lst=[]

for i in updated_productList:
    prod=i[0] + "," + i[1] + "," + i[2] + ","+ i[3] + "\n"
    lst.append(prod)
lst[-1]=lst[-1][:-1]

fd=open("Inventory.txt","w")

#updating inventory file
for i in lst:
    fd.write(i)
fd.close()

print("*******************************")
print("Inventory updated")
```
